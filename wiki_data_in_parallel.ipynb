# Concurrent import of data

```python
import concurrent.futures
import string

def get_content(filename):
    with open('wiki//{}'.format(filename)) as f:
        content = f.read()
    return content

def thread_get():
    pool = concurrent.futures.ThreadPoolExecutor(max_workers=2)
    content = pool.map(get_content, files)
    content = list(content)

    articles = [f.replace('.html','') for f in files]
    return (articles, content)
    
```

```python
articles, content = thread_get()
```

# Parsing using a pool

```python
def parse_content(content):
    soup = BeautifulSoup(content,'html.parser')
    cont = soup.find('div', attrs={'id': 'content'})
    text = cont.text
    return text

def process_parse():
    pool = concurrent.futures.ProcessPoolExecutor(max_workers=2)
    text_content = pool.map(parse_content, content)
    text_content = list(text_content)

    return text_content
```

```python
text_contents = process_parse()
```

# Finding all tag counts

```python
count = soup.find_all()
for c in count:
    print(c.name, '\n')
```

```python
def get_counts(content):
    soup = BeautifulSoup(content,'html.parser')
    count = soup.find_all()
    tag_counts = {}
    for c in count:
        if c.name in tag_counts:
            tag_counts[c.name] += 1
        else:
            tag_counts[c.name] = 1
    return tag_counts

def process_tag_get():
    pool = concurrent.futures.ProcessPoolExecutor(max_workers=2)
    tag_counts = pool.map(get_counts, content)
    tag_counts = list(tag_counts)
    results = list(zip(articles,tag_counts))

    return results
```

```python
tag_counts = process_tag_get()
```

```python
tag_counts[0]
```

```python
keys = set()
for tag in tag_counts:
    for key in tag[1]:
        keys.add(key)

total_tag_counts = {}
for key in keys:
    count = 0
    for tag in tag_counts:
        for t in tag[1]:
            if t == key:
                count += tag[1][key]
    total_tag_counts[key] = count
    
print(total_tag_counts)
```
